```python
import pandas as pd

import_og = pd.read_excel('Avon_Data.xlsx')
import_10k = pd.read_excel('Avon_Data_virtual_walker_10k.xlsx')
import_1m = pd.read_excel('Avon_Data_virtual_walker_1m.xlsx')

#original data
df_og = pd.DataFrame(import_og)
#all 392 virtual walkers got a random amount from 2k-10k
df_10k = pd.DataFrame(import_10k)
##all 392 virtual walkers got a random amount from 500k-1M
df_1m = pd.DataFrame(import_1m)
```

```python
#editing a bit for readability
df_og=df_og[df_og['Year']!= 2009]
df_og=df_og[df_og['Year']!= 2017]
df_og=df_og[['Year', 'City', 'Classification', 'Walk_Status', 'Total_Raised_Master']]

df_10k=df_10k[df_10k['Year']!= 2009]
df_10k=df_10k[df_10k['Year']!= 2017]
df_10k=df_10k[['Year', 'City', 'Classification', 'Walk_Status', 'Total_Raised_Master']]

df_1m=df_1m[df_1m['Year']!= 2009]
df_1m=df_1m[df_1m['Year']!= 2017]
df_1m=df_1m[['Year', 'City', 'Classification', 'Walk_Status', 'Total_Raised_Master']]
```

```python
#mean amount is 18$
df_og.loc[df_og['Classification'] == 'Virtual Walker'].describe()
```

```python
#mean amount for virtual walkers is $6k
df_10k.loc[df_10k['Classification'] == 'Virtual Walker'].describe()
```

```python
#mean amount for virtual walkers is $750k
df_1m.loc[df_1m['Classification'] == 'Virtual Walker'].describe()
```

```python
from ols_regression import ols_regression
from more_results import extend_results

#r-value is .14 for original data
hot_df_og=extend_results(df_og,'Total_Raised_Master', OneHotEncoder=True)
ols_regression(hot_df_og, 'Total_Raised_Master')
```

```python
#r-value is still .14 for original data
hot_df_10k=extend_results(df_10k,'Total_Raised_Master', OneHotEncoder=True)
ols_regression(hot_df_10k, 'Total_Raised_Master')
```

```python
#all r-values dropped but classification didnt and on the top
hot_df_1m=extend_results(df_1m,'Total_Raised_Master', OneHotEncoder=True)
ols_regression(hot_df_1m, 'Total_Raised_Master')
```
